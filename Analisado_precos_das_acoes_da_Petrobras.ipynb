```python
#plotando preços de fechamento da ação AAPL(apple)
import plotly.graph_objs as go 

data = [go.Scatter(
            x=df.Date,
            y=df['Close'])]
py.iplot(data)
```

```python
#configurar faixa de eixos(dados/range)
import plotly.graph_objs as go 

Close = go.Scatter(
            x=df.Date,
            y=df.Close,
            name = "PETR4 Close",
            line = dict(color = '#17BECF'),
            opacity = 0.8)
data = [Close]

layout = dict(
    title = "Faixa de Data Customizada",
    title_x = 0.5,
    xaxis = dict(
        range = ['2020-01-01', '2020-12-31'])
)

fig = dict(data=data, layout=layout)
py.iplot(fig, filename= "Manually Set Range") 
```

```python
#rangeslider e rangeselector 
import plotly.graph_objs as go 

Close = go.Scatter(
            x=df.Date,
            y=df.Close,
            name = "PETR4 High",
            line = dict(color = '#17BECF'),
            opacity = 0.8)
data = [Close]

layout = dict(
    title ="Série com Rangeslider e Botoes",
    title_x= 0.5,
    xaxis = dict(
        rangeselector=dict(
            buttons=list([
                          dict(count=1,
                               label='1m',
                               step='month',
                               stepmode='backward'),
                          dict(count=6,
                               label='6m',
                               step='month',
                               stepmode='backward'),
                          dict(step='all')
            ])
        ),
        rangeslider=dict(
            visible = True
        ),
        type='date'
    )
)

fig = dict(data= data, layout= layout)
py.iplot(fig) 
```

```python
#canclesticks , variação dos preços conforme abertura, baixa, alta, fechamento ....
trace = go.Candlestick(x=df['Date'],
                       open=df['Open'],
                       high=df['High'],
                       low=df['Low'],
                       close=df['Close'])
data = [trace]
py.iplot(data, filename='simple_candlestick')
```

```python
#sem rangeslider

trace = go.Candlestick(x=df['Date'],
                       open=df['Open'],
                       high=df['High'],
                       low=df['Low'],
                       close=df['Close'])


layout = go.Layout(
    title='Preços em Candlesticks',
    title_x=0.5,
    xaxis = dict(
        rangeslider = dict(visible = False)
    )
)

data = [trace]
fig = go.Figure(data=data, layout=layout) 
py.iplot(fig) 
```

```python
#customizando candlestick
trace = go.Candlestick(x=df['Date'],
                       open=df['Open'],
                       high=df['High'],
                       low=df['Low'],
                       close=df['Close'],
                       increasing=dict(line=dict(color= '#17BECF')),
                       decreasing=dict(line=dict(color= '#7F7F7F'))
                       )


layout = go.Layout(
    title='Preços em Candlesticks',
    title_x=0.5,
    xaxis = dict(
        rangeslider = dict(visible = False)
    )
)

data = [trace]
fig = go.Figure(data=data, layout=layout) 
py.iplot(fig) 
```

```python
#trabalhando com médias

#Média simples de 3 dias
df['MM_3'] = df.Close.rolling(window=3).mean()

#Média simples de 7 dias
df['MM_7'] = df.Close.rolling(window=7).mean()

#Média simples de 9 dias
df['MM_9'] = df.Close.rolling(window=9).mean()

#Média simples de 17 dias
df['MM_17'] = df.Close.rolling(window=17).mean()

#vizualizando as colunas criadas
df.head(20)
```

```python
#plota as média e os preços

close = go.Scatter(
    x=df.Date,
    y=df.Close,
    name="PETR4 Close",
    line=dict(color = '#330000'),
    opacity = 0.8)

MM_3 = go.Scatter(
    x=df.Date,
    y=df['MM_3'],
    name="Média Móvel 3 Períodos",
    line=dict(color = '#0000FF'),
    opacity = 0.8)

MM_7 = go.Scatter(
    x=df.Date,
    y=df['MM_7'],
    name="Média Móvel 7 Períodos",
    line=dict(color = '#FF0000'),
    opacity = 0.8)

MM_9 = go.Scatter(
    x=df.Date,
    y=df['MM_9'],
    name="Média Móvel 9 Períodos",
    line=dict(color = '#B2FF66'),
    opacity = 0.8)

MM_17 = go.Scatter(
    x=df.Date,
    y=df['MM_17'],
    name="Média Móvel 17 Períodos",
    line=dict(color = '#FF00FF'),
    opacity = 0.8)

data = [close, MM_3, MM_7, MM_9, MM_17]

py.iplot(data)
```
